```python
import pandas as pd
import pandas_datareader.data as web
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
style.use("bmh")
```

```python
start = dt.datetime(2021,1,1)
end = dt.datetime(2021,12,31)
```

```python
tick = web.DataReader('GBTC', 'yahoo', start, end)
```

```python
tick.to_csv('gbtc.csv')
```

```python
tick = pd.read_csv('gbtc.csv')
print(tick)
```

```python
pd.set_option("display.precision", 2)
tick
```

```python
first_col = tick.iloc[:,0]
print(first_col)
```

```python
print(tick[['Volume', 'Adj Close']].tail())
```

```python
tick.plot()
```

```python
close = tick['Adj Close']
close.plot()
plt.title("Title")
```

```python
tick['50ma'] = tick['Adj Close'].rolling(window=50).mean()
```

```python
tick['50ma'].plot()
plt.title("50 Day Simple moving average")
```

```python
plt.figure(figsize = (12,6))


plt.plot( label='GBTC Adj Close', linewidth = 2)
plt.plot(close, label='50 day rolling SMA', linewidth = 1.5)


plt.xlabel('Date')
plt.ylabel('Adjusted closing price ($)')

plt.title('Price with a single Simple Moving Average')
plt.legend();
```
